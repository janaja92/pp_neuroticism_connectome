# Create Simulated Neuroticism Data

```python
import pandas as pd
import numpy as np

num_rows = 1206
nums = [1, 6, 11, 16, 21, 26, 31, 36, 41, 46, 51, 56]
neoraw_cols = [f"NEORAW_{i:02d}" for i in nums]

#  Generate random NEORAW data  (values 0-4)
neoraw_data = {col: np.random.randint(0, 5, size=num_rows) for col in neoraw_cols}

#  Calculate NEOFAC_N as the sum of the NEORAW columns for each row
neofac_n = np.sum([neoraw_data[col] for col in neoraw_cols], axis=0)

# Assemble the data dictionary
data = {
    'Subject': np.random.randint(100000, 999999, size=num_rows), # Generate Subject codes
    'NEOFAC_N': neofac_n,
    'NEO-FFI_Compl': [True] * num_rows # always True
}

# Update dictionary with the NEORAW columns
data.update(neoraw_data)
df = pd.DataFrame(data)

# Convert numbers of NEORAW cols to strings
mapping_normal = {0: "SD", 1: "D", 2: "N", 3: "A", 4: "SA"}
mapping_reversed = {4: "SD", 3: "D", 2: "N", 1: "A", 0: "SA"}

nums_normal = [6, 11, 21, 26, 36, 41, 51, 56]
neoraw_cols_normal = [f"NEORAW_{i:02d}" for i in nums_normal]
nums_reversed = [1, 16, 31, 46]
neoraw_cols_reversed = [f"NEORAW_{i:02d}" for i in nums_reversed]

df[neoraw_cols_normal] = df[neoraw_cols_normal].replace(mapping_normal)
df[neoraw_cols_reversed] = df[neoraw_cols_reversed].replace(mapping_reversed)

# Final order and Save
column_order = ['Subject', 'NEOFAC_N', 'NEO-FFI_Compl'] + neoraw_cols
df = df[column_order]
df.to_csv('hcp_dummy_data.csv', index=False)

# Verification check
print(df.head())
print(f"\nDataFrame shape: {df.shape}")
```
